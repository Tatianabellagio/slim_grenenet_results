```python
import numpy as np
import pandas as pd
import os
import glob
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
pattern = os.path.join('/home/tbellagio/scratch/slim_grenenet/results', '**', 'allele_freq.csv')
allele_freq_files = glob.glob(pattern, recursive=True)
```

```python
allele_freq_founder_file = '/home/tbellagio/scratch/slim_grenenet/data/allele_freq_founder_offset.csv'

allele_freq_founder = pd.read_csv(allele_freq_founder_file)
```

```python
def calc_delta_p(allele_freq,allele_freq_founder ): 
    allele_freq_founder['chrom_pos'] = allele_freq_founder['chrom_pos'].astype(int)
    ## eliminate duplciates basically positions where the allele freq is the same in all 
    allele_freq = allele_freq.round(6)

    allele_freq = allele_freq.set_index('chrom_pos')   #.drop_duplicates()
    allele_freq = allele_freq.reset_index()
    allele_freq = allele_freq.fillna(0)

    ## first calculate delta p norm 
    ## imoport allele freq of the founder and normalize 
    allele_freq = pd.merge(allele_freq,allele_freq_founder, on ='chrom_pos')
    allele_freq = allele_freq.set_index('chrom_pos')
    
    delta_p = pd.DataFrame() #index = allele_freq.index

    for col in allele_freq.columns:
        delta_p[col] = allele_freq[col] - allele_freq['allele_freq_founder']
        
    delta_p = delta_p.drop(['allele_freq_founder','deno_norm'],axis=1)

    delta_p = delta_p.round(6)
    ## eliminate rows with all the same values
    #delta_p = delta_p[delta_p.std(axis=1) > 0]
    return delta_p
```

```python
def calc_delta_p_norm(allele_freq,allele_freq_founder ): 
    allele_freq_founder['chrom_pos'] = allele_freq_founder['chrom_pos'].astype(int)
    ## eliminate duplciates basically positions where the allele freq is the same in all 
    allele_freq = allele_freq.round(6)

    allele_freq = allele_freq.set_index('chrom_pos')   #.drop_duplicates()
    allele_freq = allele_freq.reset_index()
    allele_freq = allele_freq.fillna(0)

    ## first calculate delta p norm 
    ## imoport allele freq of the founder and normalize 
    allele_freq = pd.merge(allele_freq,allele_freq_founder, on ='chrom_pos')
    allele_freq = allele_freq.set_index('chrom_pos')
    
    delta_p = pd.DataFrame() #index = allele_freq.index

    for col in allele_freq.columns:
        delta_p[col] = (allele_freq[col] - allele_freq['allele_freq_founder']) / allele_freq['deno_norm']
        
    delta_p = delta_p.drop(['allele_freq_founder','deno_norm'],axis=1)

    delta_p = delta_p.round(6)
    ## eliminate rows with all the same values
    #delta_p = delta_p[delta_p.std(axis=1) > 0]
    return delta_p
```

```python
def df_for_plot(causal_loci, df):
    ##filter causal loci 
    #df = df[df['chrom_pos'].isin(causal_loci.pos)]
    #df = df.set_index('chrom_pos')
    ##calcualte the mean 
    #df = df.groupby(df.columns.str.split('_').str[0], axis=1).mean()
    ## order columns 
    #df = df[['optima-2', 'optima-1', 'optima0', 'optima1', 'optima2']]
    #merge with causal loci to get the selection coeffcient 
    ## by merging  i only include in the possitvie or negative and get rid of the causal loci that got lost in all the optima
    df = df.merge(causal_loci, left_on ='chrom_pos', right_on = 'pos', how='inner' )
    df = df.drop(['chrom_pos', 'pos'],axis=1)
    df = df.fillna(0)

    #df['sc'] = df['sc'].abs()
    df = df.sort_values(by='sc') #.set_index('sc')

    custom_colors = sns.color_palette("Greens", n_colors=len(df))
    # Create a custom color dictionary
    custom_color_dict = dict(zip(df['sc'], custom_colors))
    df['color'] = df['sc'].map(custom_color_dict)
    
    df['sc'] = df['sc']#.round(4)
    df = df.set_index('sc')
    
    return df
```

```python
def generate_3_df (path_allele_freq):

    allele_freq = path_allele_freq
    causal_loci_file = allele_freq.split('arq')[0] + 'arq' + allele_freq.split('arq')[1].split('/')[0] + '/loci_effectsize.csv'
    causal_loci = pd.read_csv(causal_loci_file, index_col = [0])
    pos_causal_loci = causal_loci[causal_loci['sc'] > 0]
    neg_causal_loci = causal_loci[causal_loci['sc'] < 0]

    allele_freq = pd.read_csv(allele_freq)

    allele_freq = allele_freq[allele_freq['chrom_pos'].isin(causal_loci.pos)]
    allele_freq = allele_freq.drop('Unnamed: 0',axis=1)

    delta_p = calc_delta_p(allele_freq,allele_freq_founder )
    delta_p_norm = calc_delta_p_norm(allele_freq,allele_freq_founder )
    delta_p = delta_p.reset_index()
    delta_p_norm = delta_p_norm.reset_index()

    pos_delta_p_norm_plot = df_for_plot(pos_causal_loci, delta_p_norm)
    pos_delta_p_norm_plot['name'] = 'delta_p_norm'
    pos_delta_p_norm_plot['sign'] = 'pos'
    pos_delta_p_plot = df_for_plot(pos_causal_loci, delta_p)
    pos_delta_p_plot['name'] = 'delta_p'
    pos_delta_p_plot['sign'] = 'pos'
    pos_allele_freq_plot = df_for_plot(pos_causal_loci, allele_freq)
    pos_allele_freq_plot['name'] = 'allele_freq'
    pos_allele_freq_plot['sign'] = 'pos'
    
    neg_delta_p_norm_plot = df_for_plot(neg_causal_loci, delta_p_norm)
    neg_delta_p_norm_plot['name'] = 'delta_p_norm'
    neg_delta_p_norm_plot['sign'] = 'neg'
    neg_delta_p_plot = df_for_plot(neg_causal_loci, delta_p)
    neg_delta_p_plot['name'] = 'delta_p'
    neg_delta_p_plot['sign'] = 'neg'
    neg_allele_freq_plot = df_for_plot(neg_causal_loci, allele_freq)
    neg_allele_freq_plot['name'] = 'allele_freq'
    neg_allele_freq_plot['sign'] = 'neg'
    
    
    #dict({'allele_freq':pos_allele_freq_plot, 'deltap': pos_delta_p_plot, 'deltapnorm':pos_delta_p_norm_plot})
    
    return pd.concat([pos_delta_p_norm_plot, pos_delta_p_plot,pos_allele_freq_plot, neg_delta_p_norm_plot, neg_delta_p_plot, neg_allele_freq_plot],axis=0)
```

```python
allele_freq_files = [i for i in allele_freq_files if '/strongsel' in i]
allele_freq_files = [i for i in allele_freq_files if 'highh' in i]
allele_freq_files = [i for i in allele_freq_files if 'highfreq' in i]
allele_freq_files = [i for i in allele_freq_files if '_1' in i]

files = pd.DataFrame({'file': allele_freq_files})
files['heritability'] = files['file'].str.split('/').str[-3]
files['freq'] = files['file'].str.split('/').str[-4].str.split('_').str[1]
files['pi'] = files['file'].str.split('/').str[-4].str.split('_').str[2]
files['rep'] = files['file'].str.split('/').str[-4].str.split('_').str[3]

custom_order = ['monogen', 'fivepoly',  'twentypoly', 'onehpoly']
custom_order_cat = pd.Categorical(files['pi'], categories=custom_order, ordered=True)
files['order1'] = custom_order_cat
custom_order = ['lowfreq', 'mediumfreq', 'highfreq']
custom_order_cat = pd.Categorical(files['freq'], categories=custom_order, ordered=True)
files['order2'] = custom_order_cat
custom_order = ['lowh', 'mediumh', 'highh']
custom_order_cat = pd.Categorical(files['heritability'], categories=custom_order, ordered=True)
files['order3'] = custom_order_cat


files = files.sort_values(by=['order1'])

files['file'].values



pos_arqs = []
for i in files.file:
    arq1 = generate_3_df(i)
    #name = files[files['file'] == i]['order1'].values[0]
    #pos_1_arq['arq'] = name
    pos_arqs.append(arq1)
  
```

```python
arq_monogen = pos_arqs[0].copy()

arq_monogen = arq_monogen.reset_index()

all_df = pd.melt(arq_monogen, id_vars=['name', 'sc', 'color', 'sign'], 
                    var_name='subp', value_name='value')
all_df['color'] = 'rgb' + all_df['color'].astype(str) + ''

category_orders={ # replaces default order by column name
                "name": ['allele_freq', 'delta_p', 'delta_p_norm']}
```

```python
all_df['sc'] = all_df['sc'].round(5)
```

```python
import plotly.express as px

# Assuming all_df is your DataFrame
fig = px.line(all_df, x="subp", y="value", color='sc', color_discrete_sequence=all_df['color'],
              facet_col="name", facet_row='sign',
              #facet_col_names={'allele_freq': 'Allele frequencies raw', 'delta_p': 'Delta p', 'delta_p_norm': 'Delta p normalized (pi * (1-pi))'},
              category_orders=category_orders)

# Update the y-axis range for each subplot individually
fig.update_yaxes(matches=None, autorange=True, showticklabels=True)

# Set the titles for each subplot
#fig.update_xaxes(title_text="X-axis Title")
#fig.update_yaxes(title_text="Y-axis Title")

fig.update_yaxes(title_text="Positive effect size causal loci", row=2, col=1)
fig.update_yaxes(title_text="Negative effect size causal loci", row=1, col=1)

# Set the figure size
fig.update_layout(height=1000, width=1800, autosize=True)

# Show the figure
fig.show()

    
```
