```python
## add environmental noise based on heritability and standarize 
h2s = dict()
for h2 in [0.1, 0.5, 0.9]:
    real_pheno = pd.DataFrame()
    for i in all_pheno_df_og.columns:
        va = all_pheno_df_og[i].var()
        ve = (va - h2 * va) / h2
        env_noise = np.random.normal(0, np.sqrt(ve), 231)
        real_pheno[i + '_h2' + str(h2)] = all_pheno_df_og[i] + env_noise
        
        pheno_mean = np.array(real_pheno).mean()
        pheno_std = np.array(real_pheno).std()
        phenotypes_st = (real_pheno - pheno_mean)/ pheno_std
    h2s[str(h2)] = phenotypes_st
```

```python
phenotypes_st
```

```python
real_phenos = pd.concat(h2s.values(), axis=1)
```

```python
all_pheno_df = real_phenos.T.reset_index()
```

```python
all_pheno_df['index'] = all_pheno_df['index'].str.split('arq_').str[1]

all_pheno_df['freq'] = all_pheno_df['index'].str.split('_').str[0]

all_pheno_df['rep'] = all_pheno_df['index'].str.split('_').str[2].str.split('/').str[0]
all_pheno_df['poly'] = all_pheno_df['index'].str.split('_').str[1]

all_pheno_df['h2'] = all_pheno_df['index'].str.split('h2').str[1]

all_pheno_df = all_pheno_df.drop('index',axis=1)

#all_pheno_df

#all_pheno_df = all_pheno_df.pivot_table(columns = ['poly','freq', 'h2', ''],values=all_pheno_df.columns[:-3])


#all_pheno_df = all_pheno_df.melt()

all_pheno_df = all_pheno_df.melt(id_vars = ['poly', 'freq', 'h2', 'rep']).drop('variable',axis=1)
```

```python
def arrange_dataset(files):
    joint = pd.DataFrame()
    for i in files:
        one_file =pd.read_csv(i,header=None)
        one_file.columns = [i]
        joint = pd.concat([joint, one_file],axis=1)
    joint = joint.T
    joint.index = joint.index.str.split('/subp').str[0]
    #joint_meansubp = joint.reset_index().groupby('index')[joint.columns].mean()
    return joint
```

```python
# Use the glob module to search for files in subfolders
pattern = os.path.join(path2, '**', '*_mfitness.txt')
mfitness_files = glob.glob(pattern, recursive=True)

mfitness = arrange_dataset(mfitness_files)

mfitness.to_csv('mfitness_oct4.csv')
```
